```python
#importing relevant libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn as sk
import scipy as sc
%matplotlib inline
```

```python
#Importing the dataset
frame = pd.read_csv("Data/taxi.csv")
frame
```

```python
#variable 1
x = frame["passenger_count"]
x
```

```python
y = frame.loc[:,"trip_distance"]
y
```

```python
plt.scatter(x, y)
plt.xlabel("passenger_count")
plt.ylabel ("trip_distance")
plt.title("Implementing a simple linear regression on passenger_count and trip_distance")
plt.show()

```

```python
from sklearn.linear_model import LinearRegression
```

```python
#defining the class
model=LinearRegression(fit_intercept=True)
model
```

```python
#mulptiple linear regression
y=y[:,np.newaxis]
y
```

```python
x=x[:,np.newaxis]
x
```

```python
model.fit(x,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
xfit = np.linspace(-1,10)
xfit = xfit[:,np.newaxis]
xfit
```

```python
yfit = model.predict(xfit)
yfit
```
